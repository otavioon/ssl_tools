```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leia o arquivo CSV
nome_arquivo = "/workspaces/betania/ssl_tools/test/1.csv"
dados = pd.read_csv(nome_arquivo)

# Examine o formato do arquivo CSV para determinar os nomes das colunas
# Suponha que as colunas sejam nomeadas 'accel_x', 'accel_y', 'accel_z', 'gyro_x', 'gyro_y', 'gyro_z'
# Se os nomes das colunas forem diferentes, ajuste o cdigo de acordo.

# Plotando os dados do acelermetro

plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
plt.plot(dados['accel-x'], label='Acelermetro X')
plt.plot(dados['accel-y'], label='Acelermetro Y')
plt.plot(dados['accel-z'], label='Acelermetro Z')
plt.plot(dados['activity code'], label='Atividade')
plt.title('Dados do Acelermetro')
plt.xlabel('Amostras')
plt.ylabel('Valores')
plt.yticks(range(-10, 13, 1))
plt.legend()
plt.grid(True)
plt.xlim([0, len(dados)]) 

#plt.xlim([0, 5000]) 

# Plotando os dados do giroscpio
plt.subplot(2, 1, 2)
plt.plot(dados['gyro-x'], label='Giroscpio X')
plt.plot(dados['gyro-y'], label='Giroscpio Y')
plt.plot(dados['gyro-z'], label='Giroscpio Z')
plt.title('Dados do Giroscpio')
plt.xlabel('Amostras')
plt.ylabel('Valores')
plt.legend()
plt.grid(True)

# Ajuste os limites do eixo X conforme necessrio
plt.xlim([0, len(dados)]) 

plt.savefig('plot_user1.png')

# Mostrar o grfico
plt.tight_layout()
plt.show()



```
